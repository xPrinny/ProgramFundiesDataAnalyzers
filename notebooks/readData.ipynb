#### 3. Data Exploration

##### a) Channel Id

```python
# no.of unique channel ID
gamingData['channelId']nunique()
```

```python
# check for null
gamingData['channelId'].isnull().sum()
```

```python
# top 5 channel IDs by frequency
gamingData['channelId'].value_counts()[:5].sort_values()
```

##### b) Channel Title

```python
# no.of unique channel Title
gamingData['channelTitle'].nunique()
```

```python
# top 5 channel Title by frequency
gamingData['channelTitle'].value_counts()[:5].sort_values()
```

##### c) Trending Date

```python
# Top 5 trending dates by frequency
gamingData['trending_date'].value_counts()[:5].sort_values()
```
