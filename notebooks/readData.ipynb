##### d) Channels with same ID but different name

```python
# Loops through channelTitle and make it a list, insert values into a temporary Data Frame for each channelTitle for that channelID.
channelIdGroup = []
for col in ['channelTitle']:
    temp_df = gamingData.groupby('channelId')[col].apply(lambda x : pd.Series(x.unique().tolist())).unstack()
    temp_df.columns = temp_df.columns.map(lambda x : f"{col}_{x+1}")
    channelIdGroup.append(temp_df)
channelIdGroup = pd.concat(channelIdGroup)

pd.DataFrame(channelIdGroup[channelIdGroup['channelTitle_2'].notna()], columns=["channelId", "channelTitle_1", "channelTitle_2"])
```

##### e) Filtering videoes to required data (Total videos, View count, Likes, Dislikes, Average view per Video, Like and Dislike ratio)

```python
# Remove all duplicate video and only keep the latest from the trend
newChannelData = gamingData.sort_values('trending_date').drop_duplicates('title',keep='last')

# Get total number of Appereance for a video
newChannelData = newChannelData.merge(gamingData.groupby(['title']).size().reset_index(), how='outer', left_on=['title'], right_on=['title']).rename(columns={0:'Total Apperance'})

# Groups by 'channelId', totaling the number of videos the channel produced, total view counts, likes, and dislikes
channelTotalData = newChannelData.groupby('channelId').agg({'channelTitle':'max', 'channelId':'count', 'view_count': 'sum', 'likes': 'sum', 'dislikes': 'sum'}).rename(columns={'channelId':'Total Videos'}).reset_index()

# Average view per video
channelTotalData['avgViewPerVideo'] = (channelTotalData['view_count'] / channelTotalData['Total Videos'])

# Likes to dislike ratio
channelTotalData['ratioLikeDislike'] = (channelTotalData['likes'] / (channelTotalData['likes'] + channelTotalData['dislikes']))
```

#### 4. Channel Anaylsis

##### 4.1 Top 20 channel with most overall views


```python
# Gets top 20 channels based on total video count
channelTotalData.sort_values(by=['view_count'], ascending=False).head(20)
```

##### a) View count vs Amount of days the video has been trending

```python
# Stores the top 20 channels based on total video count into a list
top20channelIDView = channelTotalData.sort_values(by=['view_count'], ascending=False).head(20)['channelId'].tolist()

# Declaring new Series
top20channelIDViewX = pd.Series([], dtype='float64')
top20channelIDViewY = pd.Series([], dtype='float64')

# Fill the new Series with top 20 channel view count and Total Appearance
for ids in top20channelIDView:
    top20channelIDViewX = pd.concat([top20channelIDViewX, newChannelData[newChannelData['channelId'] == ids]['view_count']])
    top20channelIDViewY = pd.concat([top20channelIDViewY, newChannelData[newChannelData['channelId'] == ids]['Total Apperance']])

# Divide the views by 1 million cause the numbers are too big
top20channelIDViewX = top20channelIDViewX.divide(1000000)

# Plot scatterplot and Linear Regression
plt.scatter(top20channelIDViewX, top20channelIDViewY)
reg = np.polyfit(top20channelIDViewX, top20channelIDViewY, deg = 1)
trend = np.polyval(reg, top20channelIDViewX)
plt.plot(top20channelIDViewX, trend, 'r')
plt.xlabel("View Count (Million)")
plt.ylabel("Days trending")

plt.show()
```

```python
# Plot again and remove outlier
plt.scatter(top20channelIDViewX, top20channelIDViewY)
reg = np.polyfit(top20channelIDViewX, top20channelIDViewY, deg = 1)
trend = np.polyval(reg, top20channelIDViewX)
plt.plot(top20channelIDViewX, trend, 'r')
plt.xlabel("View Count (Million)")
plt.ylabel("Days trending")
plt.ylim(0,12)
plt.xlim(0,20)

plt.show()
```
